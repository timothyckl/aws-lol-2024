1. Find all duplicated questions using cosine similarity.
2. Remove duplicated questions from the training set keeping the one with the longest response

```python
import json
import numpy as np
import pandas as pd

# read the array from the file
with open('joel-dataset-2.json', 'r') as file:
    data = json.load(file)


# flatten the array
flattened_list = [item for sublist in data for item in sublist]
# convert the array to a dataframe
df = pd.DataFrame(flattened_list)
# rename question to instruction and answer to response
df = df.rename(columns={'question': 'instruction', 'answer': 'response'})

# random shuffle the dataframe
df = df.sample(frac=1).reset_index(drop=True)
df.to_json('joel-partyrock-1.jsonl', orient='records', lines=True)
```
