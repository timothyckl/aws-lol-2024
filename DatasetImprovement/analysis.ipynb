```python
import pandas as pd

# read jsonl file
dataset_7 = pd.read_json('../notebooks/dataset-7.jsonl', lines=True)
dataset_7
# create a histogram of the length of response
dataset_7['response'].str.len().hist()
# find the longest response and its index ( longest in characters)
longest_response = dataset_7['response'].str.len().idxmax()
print('longest_response index is ', longest_response)
print('num of rows in dataset_8 is ', dataset_7.shape[0])

```
